# Extract Date from the data

```python
import os
import pandas as pd
import sqlite3
```

```python
input_file = 'final.db'
table = '\"1- Patents\"'
conn = sqlite3.connect(input_file)
```

```python
df = pd.read_sql_query("SELECT * FROM \"1- Patents\"",conn)
search_list = ['Publication Date', 'Application Date']
print(df[search_list].head())
```

```python
def extract_year(filename, df, search_list):
    # open csv file
    for name in search_list:
        print(df[name])
        # extract year from date which has format of mm/dd/yyyy
        df[name + ' Year'] = df[name].apply(lambda x: x.split('/')[0])
        print(df[name + ' Year'])
    
    # save to new csv file
    df.to_sql('1- Patents', conn, if_exists='replace', index=False)
```

```python
def reformat_date(filename, df, search_list):
    # open csv file
    for name in search_list:
        # extract year from date which has format of mm/dd/yyyy
        for item in range(0, df[name].size):
            temp = df[name][item].split('/')
            if len(temp[1]) == 1:
                temp[1] = '0' + temp[1]
            if len(temp[2]) == 1:
                temp[2] = '0' + temp[2]
            df[name][item] = temp[0] + '/' + temp[1] + '/' + temp[2]
        
        print(df[name])
            
    #save to sql
    df.to_sql('1- Patents', conn, if_exists='replace', index=False)
```

## Driver code

### Extract Year Driver

```python
extract_year(input_file, df, search_list)
conn.close()
```

### Reformat Date Driver

```python
reformat_date(input_file, df, search_list)
conn.close()
```
